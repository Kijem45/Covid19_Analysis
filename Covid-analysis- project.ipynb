```python
#To check the countries that are not appropriate or not correct/remove the inappropriate ones
df["location"].isin(["Africa", "World", "Upper middle income","Low income", "Lower middle income"])
```

```python
condition = df["location"].isin(["Africa", "World", "Upper middle income", "Low income", "Lower middle income"])
df.drop(df[condition].index, inplace=True)    #drop the index number that is true/ try not to run this line of code twice
df["location"].unique()
```

```python
#To start from the recent ones/recent cases
#most_recent=df[df["date"] == max(df["date"])].reset_index()  # This will reset the index to match your sorted values
#most_recent=df[df["date"] == max(df["date"])]
#most_recent
#df.sort_values(by="date", ascending=False)
most_recent=df.sort_values(by="date", ascending=False)
most_recent
```

```python
#To convert our date datatype to type date
type(df["date"])
```

```python
type(df["date"][0])   #Check the index
```

```python
most_recent["date"] = pd.to_datetime(most_recent["date"]).dt.strftime("%y/%m/%d")
most_recent
```

```python
#To set the date as the index
#most_recent.reset_index("date")
most_recent.set_index("date")
most_recent = most_recent.set_index("date")   #To save it
most_recent
```

```python
#Save our data for flourish- flourish is the website we will do our visualizatio/use flourish as a variable
flourish_data = pd.pivot_table(most_recent, values="total_cases", index=["location"], columns="date").reset_index()
flourish_data
```

```python
flags = pd.read_csv("flags_world.csv")
flags
```

```python
#Merging the data
#final_data = pd.merge(flourish_data, flags, how="country") this is if flags have country in common with covid data 
final_data = pd.merge(flourish_data, flags, left_on="location", right_on="country")
final_data.head()
```

```python
#This drops country data
#del final_data["country"]   removing country column
#Exporting our data
img = final_data["Image URL"] #Bringing out the image/backing up image url column
final_data.drop(labels=["Image URL"], axis=1, inplace=True)   #removing image url from last column
final_data.insert(1,"flags",img)  #inserting into index 1/inserting image url to second column
final_data.to_csv("covid_final.csv") #saving  final data to csv
final_data
```
